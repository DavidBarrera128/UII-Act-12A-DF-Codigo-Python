```python
# Programa para calcular el área de un triángulo

# Solicitamos la base y la altura asegurando que sean números enteros
base = int(input("Introduce la base del triángulo (número entero): "))
altura = int(input("Introduce la altura del triángulo (número entero): "))

# Aplicamos la fórmula: (Base * Altura) / 2
# Usamos un punto para representar la división y obtener un resultado preciso
area = (base * altura) / 2

# Mostramos el resultado
print(f"El área del triángulo con base {base} y altura {altura} es: {area}")
```
